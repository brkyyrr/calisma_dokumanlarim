**<b><center>MATPLOTLIB</center></b>**

&nbsp; &nbsp; &nbsp; &nbsp;Verileri görselleştirip, 2 ve 3 boyutlu çizimler yapmamızı sağlar.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;Şimdi TÜFE verisini grafik üzerinden inceleyelim. 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel("tüfe.xls")
```

```python
data.index=["2003","2004","2005","2006","2007","2008",
            "2009","2010","2011","2012","2013","2014",
            "2015","2016","2017","2018","2019","2020",
            "2021","2022","2023"]
data.drop("Yıl",axis=1,inplace=True)
```

```python
#data["Ocak"] = data["Ocak"].astype("float")
#data["Şubat"] = data["Şubat"].astype("float")
```

```python
data.head()
```

*<i><b>Temel Görselleştirme ve Genel Komutlar</b></i><br/>*
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.plot()</b> grafik oluşturmanın temel komutudur.Grafiği tanımladıktan sonra, özelliklerini ekleriz.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.figure(figsize=())</b>, grafiğin boyutunu belirler. x eni, y ise boyudur.   <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.xlabel()</b>, x eksesine ait başlık bilgisi<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.ylabel()</b>, y eksesine ait başlık bilgisi<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.title()</b>, grafiğin başlık bilgisidir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.show()</b>, grafiğin çizilmesini ve ekranda gözükmesini sağlar.<br/>

```python
plt.figure(figsize=(12,10))
plt.plot(data.index, data["Şubat"])
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.show()
```

&nbsp; &nbsp; &nbsp; &nbsp;İki farklı veriyi tek grafik üzerinde gösterelim. Burada dikkat etmemiz gereken konu, iki verinin de x veya y eksenlerinin ortak olmasıdır.

```python
plt.figure(figsize=(12,10))
plt.plot(data.index, data["Şubat"], color="red")
plt.plot(data.index, data["Mart"], color="blue")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.show()
```

&nbsp; &nbsp; &nbsp; &nbsp;<b>subplot(x,y,z)</b> metodu birden fazla grafiği aynı anda çizmemizi sağlar. x satır, y sütun ve z ise kaçıncı grafik olduğunu ifade eder.

```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.plot(data.index, data["Ocak"], color="red")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")

plt.subplot(2,2,2)
plt.plot(data.index, data["Şubat"], color="blue")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.show()
```

&nbsp; &nbsp; &nbsp; &nbsp;<b><i>Figür Tanımlamak ve Çizmek</i></b> </br>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.figure()</b>, komutu ile de grafik çizebiliriz. Figure sayesinde tablo içerisinde çizimler yapabiliriz.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>.add_axes()</b>, koordinatları belirlememizi sağlar.<br/>
İlk değer soldan ne kadar içerde başlayacağı,ikinci değer alttan ne kadar yüksekte olacağı,üçüncü değer ne kadar uzunlukta olması gerektiği,dördüncü değer ise ne kadar yükseklikte olması gerektiğini belirtmektedir. 0 ile 1 arasında değer almalıdır. 

```python
fig = plt.figure(figsize =(12,10))
axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(data.index, data["Mart"])
axes.set_xlabel("Yıl")
axes.set_ylabel("TÜFE")
axes.set_title("TÜFE Oranları")
plt.show()
```

&nbsp; &nbsp; &nbsp; &nbsp;<b>İç İçe Grafik Çizme</b> </br>
&nbsp; &nbsp; &nbsp; &nbsp;Farklı verileri karşılaştırmak ve daha iyi incelemek için iç içe grafikler faydalı olmaktadır. Bunun için 2.axes'i oluşturmamız gerekir. Burada dikkat edilmesi gereken nokta grafiklerimizin koordinatlarıdır.

```python
fig = plt.figure(figsize =(12,10))

axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.3,0.4,0.5,0.4])

axes1.plot(data.index, data["Şubat"],color="red")
axes2.plot(data.index, data["Mart"],color="blue")

axes1.set_xlabel("Yıl")
axes2.set_xlabel("Yıl")

axes1.set_ylabel("TÜFE")
axes2.set_ylabel("TÜFE")

axes1.set_title("TÜFE Oranları Şubat")
axes2.set_title("TÜFE Oranları Mart")

plt.show()
```

*<i><b>Grafiklerin Özelleştirilmesi</b></i><br/>*
&nbsp; &nbsp; &nbsp; &nbsp;Grafiği çizdikten sonra aşağıda bulunan parametreler ile farklı özellikler tanımlayabiliriz.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>color</b>, grafiğin rengini belirler. Farklı renk paletleri yazılabilir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>linewidth</b>, grafik çizgisinin kalınlığını belirler.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>linestyle</b>, çizginin stilini belirler.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>marker</b>, her kırılım noktasına belirlediğimiz şekli koyar.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markersize</b>, marker boyutunu belirler.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markerfacecolor</b>, rengini değiştirir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markeredgewidth</b>, çizgisini kalınlaştırır.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markeredgecolor</b>, çizgisinin dış rengini değiştirir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>alpha</b>, grafiğin saydamlığını değiştirir. 0'a yaklaştıkça saydamlaşır. 1'e yaklaştıkça belirginleşir.<br/>

```python
plt.figure(figsize=(12,10))
plt.plot(data.index, data["Şubat"],color="red",linewidth=3,
        linestyle="--",marker="o",markersize=9,
        markerfacecolor="yellow",markeredgewidth=5,
        markeredgecolor="blue",alpha=0.5)
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.show()
```
