**<b><center>MATPLOTLIB</center></b>**

&nbsp; &nbsp; &nbsp; &nbsp;Verileri görselleştirip, 2 ve 3 boyutlu çizimler yapmamızı sağlar.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;Şimdi TÜFE verisini grafik üzerinden inceleyelim. 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel("tüfe.xls")
```

```python
data.index=["2003","2004","2005","2006","2007","2008",
            "2009","2010","2011","2012","2013","2014",
            "2015","2016","2017","2018","2019","2020",
            "2021","2022","2023"]
data.drop("Yıl",axis=1,inplace=True)
```

```python
#data["Ocak"] = data["Ocak"].astype("float")
#data["Şubat"] = data["Şubat"].astype("float")
```

```python
data.head()
```

*<i><b>Temel Görselleştirme ve Genel Komutlar</b></i><br/>*
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.plot()</b> grafik oluşturmanın temel komutudur.Grafiği tanımladıktan sonra, özelliklerini ekleriz.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.figure(figsize=())</b>, grafiğin boyutunu belirler. x eni, y ise boyudur.   <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.xlabel()</b>, x eksesine ait başlık bilgisi<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.ylabel()</b>, y eksesine ait başlık bilgisi<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.title()</b>, grafiğin başlık bilgisidir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.show()</b>, grafiğin çizilmesini ve ekranda gözükmesini sağlar.<br/>

```python
plt.figure(figsize=(12,10))
plt.plot(data.index, data["Şubat"])
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.show()
```

&nbsp; &nbsp; &nbsp; &nbsp;İki farklı veriyi tek grafik üzerinde gösterelim. Burada dikkat etmemiz gereken konu, iki verinin de x veya y eksenlerinin ortak olmasıdır.

```python
plt.figure(figsize=(12,10))
plt.plot(data.index, data["Şubat"], color="red")
plt.plot(data.index, data["Mart"], color="blue")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.show()
```

&nbsp; &nbsp; &nbsp; &nbsp;<b>subplot(x,y,z)</b> metodu birden fazla grafiği aynı anda çizmemizi sağlar. x satır, y sütun ve z ise kaçıncı grafik olduğunu ifade eder.

```python
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.plot(data.index, data["Ocak"], color="red")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")

plt.subplot(2,2,2)
plt.plot(data.index, data["Şubat"], color="blue")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.show()
```

&nbsp; &nbsp; &nbsp; &nbsp;<b><i>Figür Tanımlamak ve Çizmek</i></b><br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.figure()</b>, komutu ile de grafik çizebiliriz. Figure sayesinde tablo içerisinde çizimler yapabiliriz.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>.add_axes()</b>, koordinatları belirlememizi sağlar.<br/>
İlk değer soldan ne kadar içerde başlayacağı,ikinci değer alttan ne kadar yüksekte olacağı,üçüncü değer ne kadar uzunlukta olması gerektiği,dördüncü değer ise ne kadar yükseklikte olması gerektiğini belirtmektedir. 0 ile 1 arasında değer almalıdır. 

```python
fig = plt.figure(figsize =(12,10))
axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(data.index, data["Mart"])
axes.set_xlabel("Yıl")
axes.set_ylabel("TÜFE")
axes.set_title("TÜFE Oranları")
plt.show()
```

<i><b>İç İçe Grafik Çizme</b></i><br/>
&nbsp; &nbsp; &nbsp; &nbsp;Farklı verileri karşılaştırmak ve daha iyi incelemek için iç içe grafikler faydalı olmaktadır. Bunun için 2.axes'i oluşturmamız gerekir. Burada dikkat edilmesi gereken nokta grafiklerimizin koordinatlarıdır.

```python
fig = plt.figure(figsize =(12,10))

axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.3,0.4,0.5,0.4])

axes1.plot(data.index, data["Şubat"],color="red")
axes2.plot(data.index, data["Mart"],color="blue")

axes1.set_xlabel("Yıl")
axes2.set_xlabel("Yıl")

axes1.set_ylabel("TÜFE")
axes2.set_ylabel("TÜFE")

axes1.set_title("TÜFE Oranları Şubat")
axes2.set_title("TÜFE Oranları Mart")

plt.show()
```

*<i><b>Grafiklerin Özelleştirilmesi</b></i><br/>*
&nbsp; &nbsp; &nbsp; &nbsp;Grafiği çizdikten sonra aşağıda bulunan parametreler ile farklı özellikler tanımlayabiliriz.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>color</b>, grafiğin rengini belirler. Farklı renk paletleri yazılabilir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>linewidth</b>, grafik çizgisinin kalınlığını belirler.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>linestyle</b>, çizginin stilini belirler.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>marker</b>, her kırılım noktasına belirlediğimiz şekli koyar.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markersize</b>, marker boyutunu belirler.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markerfacecolor</b>, rengini değiştirir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markeredgewidth</b>, çizgisini kalınlaştırır.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markeredgecolor</b>, çizgisinin dış rengini değiştirir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>alpha</b>, grafiğin saydamlığını değiştirir. 0'a yaklaştıkça saydamlaşır. 1'e yaklaştıkça belirginleşir.<br/>

```python
plt.figure(figsize=(12,10))
plt.plot(data.index, data["Şubat"],color="red",linewidth=3,
        linestyle="--",marker="o",markersize=9,
        markerfacecolor="yellow",markeredgewidth=5,
        markeredgecolor="blue",alpha=0.5)
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.show()
```

&nbsp; &nbsp; &nbsp; &nbsp;Önceki çalışmada yaptığımız gibi, iç içe grafiklerin her birinde bu özellikleri kullanabiliriz.

*<i><b>Grafik Üzerine Yazı Eklenmesi</b></i><br/>*
&nbsp; &nbsp; &nbsp; &nbsp;<b>label ve legend</b>, grafiklerin hangi değişkenlere ait olduğunu gösterir. İkisini aynı anda kullanmalısınız. <b>legend(loc=2)</b> şeklinde değer alabilir. Değer 0-10 arasında olmalıdır ve 0 en iyi yeri ifade eder. <br/>
<br>
&nbsp; &nbsp; &nbsp; &nbsp;<b>bbox_to_anchor=(0.5,0.,0.5,0.5)</b>, direkt ledengin konumunu belirler.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>fontsize</b>, Font boyut ayarıdır. Aldığı değerler, 'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large' dır. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>numpoints</b>, legendların sayısını arttırır. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markerscale</b>, legendların boyutunu değiştirmemizi sağlar. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>markerfirst</b>, Legendi başa, göstergesini sona alır. True ya da False değer alır. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>shadow</b>, legendin etrafındaki çizgilere gölge verir. True ya da False ile kullanılır. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>framealpha</b>, çizgilerinin saydamlığını ayarlar. 0-1 arasında değer alır. 1'e yaklaştıkça saydamlık artar. 0'a yaklaştıkça azalır. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>facecolor</b>, arka planına renk vermeye yarar. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>edgecolor</b>, çizgisinin renklendirmeyi sağlar.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>mode</b>, legend kutucuğunun uzunluğunu belirler. <b>expand</b> ya da <b>None</b> değerini alır. <b>expand</b> değerini verdiğimizde grafik uzunluğunca uzar. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>title</b>, başlık vermeye yarar.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>title_fontsize</b>, aşlığının boyutunu belirler. Title ile birlikte kullanılır. ınteger değerler alır.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>borderpad</b>, legend çerçevesinin boyutunu büyütür. Float değerler alır. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>handlelength</b>, kollarının uzunluğunu belirler. Float değerler alır.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>handletextpad</b>, legend noktası ile yazısı arasındaki boşluğu ayarlamamıza yarar.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>borderaxespad</b>, legend ile grafik çizgileri arasındaki boşlukları ayarlamamıza yarar. Float değerler alır.<br/>

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Şubat"],color="red",linewidth=3,
        linestyle="--",marker="o",markerfacecolor="yellow",
        markeredgewidth=5,markeredgecolor="blue")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.legend(["Şubat"],bbox_to_anchor=(0.1, 0., 0.1, 0.5))
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Şubat"],color="red",linewidth=3,
        linestyle="--",marker="o",markerfacecolor="yellow",
        markeredgewidth=5,markeredgecolor="blue")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.legend(["Şubat"],loc=5,fontsize="medium",numpoints=2)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Şubat"],color="red",linewidth=3,
        linestyle="--",marker="o",markerfacecolor="yellow",
        markeredgewidth=5,markeredgecolor="blue")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.legend(["Şubat"],loc=5,fontsize="medium",
           markerscale=2,markerfirst=False,edgecolor="r")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Şubat"],color="red",linewidth=3,
        linestyle="--",marker="o",markerfacecolor="yellow",
        markeredgewidth=5,markeredgecolor="blue")
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.legend(["Şubat"],loc=2,fontsize="medium",
           title_fontsize=16,handlelength=6)
plt.show()
```

&nbsp; &nbsp; &nbsp; &nbsp;Farklı verileri, tek legend içinde gösterebiliriz.

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Ocak"],color="blue",linewidth=3,
        linestyle="--",marker="o",markerfacecolor="white",
        markeredgewidth=5,markeredgecolor="red")

plt.plot(data.index, data["Şubat"],color="red",linewidth=3,
        linestyle="--",marker="o",markerfacecolor="yellow",
        markeredgewidth=5,markeredgecolor="blue")

plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.legend(["Ocak","Şubat"],loc=2,fontsize="medium",
           markerscale=2,edgecolor="black",
           title="Oran",title_fontsize=12)
plt.show()
```

*<i><b>Grafiğe Izgara Eklenmesi</b></i><br/>*
&nbsp; &nbsp; &nbsp; &nbsp;<b>bgrid</b>, grafiğin arkasına grid(ızgara) çizer.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>alpha</b>, grid çizgilerinin saydamlığını ayarlar.0-1 arasında değer alır. 1'e yaklaştıkça saydamlık artar.0'a yaklaştıkça azalır.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>linestyle</b>, grid çizgilerinin biçimini belirler. <b>- , -- , -. , :</b> , gibi değerler alabilir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>linewidth</b>, grid çubuklarının kalınlığını belirler. İnteger değerler alır.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>color</b>, grid çizgilerinin rengini belirler.<br/>

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Şubat"])
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.grid(alpha=0.4,linestyle="--",linewidth=2,color="y")
plt.show()
```

*<i><b>Grafiğe Üzerine Yazı Eklemek</b></i><br/>*
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.text()</b>, grafik üzerine yazı eklememizi sağlar.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.text(x,y)</b>, yazının x ve y eksenine göre koordinatlarını belirtmemiz gerekir. <br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.text(x,y,s="")</b>, yazının belirtilmesidir.<br/>
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.text(x,y,s="",fontdict={"color":"r","fontsize":35})</b>, yazı ile ilgili düzeltmeler yapmamızı sağlar. Sözlük formatında kullanılır. <br/>

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Şubat"])
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.text(x=15,y=600,s="Kırılma Noktası",fontdict={"color":"b","fontsize":16})
plt.show()
```

*<i><b>Grafiğin Kayıt Edilmesi</b></i><br/>*
&nbsp; &nbsp; &nbsp; &nbsp;<b>plt.savefig()</b>, oluşturduğumuz grafiği kaydetmemizi sağlar.<br/>

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Şubat"])
plt.title("TÜFE")
plt.xlabel("Yıl")
plt.ylabel("TÜFE Oranları")
plt.savefig("Şubat_Tüfe.png")
plt.show()
```
